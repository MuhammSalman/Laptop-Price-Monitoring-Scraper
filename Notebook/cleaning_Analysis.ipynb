<b><h1>Import Libraries</h1></b>

```python
import numpy as np
import pandas as pd
```

<b><h1>Load Dataset</h1></b>

```python
df = pd.read_csv("../Data/raw_data.csv")
```

```python
df.shape
```

```python
df.head()
```

<b><h1>Data Cleaning</h1></b>

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python

df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
# Extracting Brand name from Name column
df["Brand"]=df["Name"].str.split().str[0]
```

```python
# Removing symobl and comma from Price 
df["Price"]=df["Price"].str.replace("₹", "").str.replace(",", "").astype(int)
```

```python
# Extracting only ram from Ram column
df["Ram"]=df["Ram"].str.split().str[:2].str.join(" ")
```

```python
# Extracting only ssd from SSD column
df["SSD"]=df["SSD"].str.split().str[:2].str.join(" ")
```

```python
# Extracting processor name from Processor column
df["Processor"]=df["Processor"].str.split().str[:2].str.join(" ")
```

```python
# Extracting only size in cm from Display column
df["Display"]=df["Display"].str.split().str[:2].str.join(" ")
```

```python
# Extracting model name from Name column
df["Model"]=df["Name"].str.split().str[1]
```

```python
# Ordering the columns
df = df[['Brand', 'Model', 'Processor', 'Ram', 'SSD', 'Price', 'Display']]
```

```python
df.head()
```

```python
df.info()
```

<b><h1>Data Analysis</h1></b>

```python
# Which brand has highest price 
brand_price = df.groupby("Brand")["Price"].max().sort_values(ascending=False)
brand_price
```

```python
# which processor has highest price
proc_price=df.groupby("Processor")["Price"].max().sort_values(ascending=False)
proc_price
```

```python
# is ssd depond on price
ssd_price=df.groupby("SSD")["Price"].max().sort_values(ascending=False)
ssd_price
```

```python
# is ram depond on price
ram_price=df.groupby("Ram")["Price"].max().sort_values(ascending=False)
ram_price
```

```python
# is screensize depond on price
display_price=df.groupby("Display")["Price"].max().sort_values(ascending=False)
display_price
```

<b><h1>Save Cleaned Data</h1></b>

```python

df.to_csv("../Data/cleaned_data.csv", index=False)
print("Successfully Saved")
```
